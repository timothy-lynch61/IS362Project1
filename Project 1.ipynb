```python
#Import Pandas
import pandas as pd

# Read CSV into pandas
csv= pd.read_csv('airline.csv')

#Showing CSV without print function
csv
```

__The information above is nice, but let's make some adjustments below to make it more user friendly.__

```python
#Sort Pandas Index

sorted_index = csv.set_index(['Airline','City'])

#Showing 
sorted_index
```

__I would like to know which city has the highest percentage of delayed flights. In order to find the average we need to know the total number of flights per city. Then create a column for the percentage.__

```python
# Copy Data to new variable
working_csv = sorted_index.copy()
```

```python
#Create New Column called Total Flights

working_csv['Total Flights'] = working_csv['On time'] + working_csv['Delayed']

working_csv
```

```python
#Create New Column called Average Delay

working_csv['Average Delay'] = (working_csv['Delayed'] / working_csv['Total Flights']) * 100
working_csv
```

__As we can see the Airline/City with the lowest average delay is Alaska airlines flying from Phoenix at 5%. The longest delay is AM West in San Francisco at 28%.__
